## binderの構築方法

1. public リポジトリを作成
2. README.mdを追加し、以下を記述、push

``` 
[![Binder](https://mybinder.org/badge_logo.svg)](https://mybinder.org/v2/gh/satoshishi/python-morimori.git/main)
```

3. binderのページから、リポジトリのhttpリンクとターゲットブランチを設定し、launchを実行

## moduleの追加方法


1. requirements.txtというファイルを作成
2. 追加したいmodule名を記述
