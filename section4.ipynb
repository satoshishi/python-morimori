## 4 if

```python
val = 10

if val >= 5:
    print("5より大きい")
if val < 0:
    print("マイナス値")
elif val in (0,4,10):
    print("any")
else:
    print("other")
```

# range

```python
a = range(5)
a_res = ""

for i in a:
    a_res += str(i)
print(a_res)

b = range(5,10)
b_res = ""

for i in b:
    b_res += str(i)
print(b_res)
    
c = range(4,10,2)
c_res = ""
for i in c:
    c_res += str(i)
print(c_res)

print(sum(range(4)))

hoge = list(range(4))
print(hoge)
```

## break

```python
for h in hoge:
    if h == "yamaguti":
        break
    print(h)
```

## def

```python
val = 1

def call():
    val = 2
    print(val)
    
#関数の代入ができる
function =  call
function()
```

```python
gb = "hoge"
def global_parameter(val = gb):
    print(gb)
    
def list_parameter(val,l = None):
    if l is None:
        l = []
    l.append(val)
    return l

def multi_parameter(val1,val2 = "fujisawa",val3 = 1):
    print(val1,val2,val3)
    
multi_parameter(1)
multi_parameter(2,"satoshi")
multi_parameter(3,"hoge",4)
global_parameter()
print(list_parameter("hogematu"))
name = list_parameter("fujisawa",[])
print(list_parameter("satoshi",name))
```
